#### Carregamento e tratamento dos dados

```python
data = pd.read_excel('coronarias_1.xlsx')
```

```python
data.replace('.', np.nan, inplace=True)
data['ALTURA'] = data['ALTURA'].str.replace(',', '.').astype(float)
data['PESO'] = data['PESO'].str.replace(',', '.').astype(float)
data['IMC'] = data['IMC'].str.replace(',', '.').astype(float)
data.fillna(data.median(), inplace=True)
```

```python
data['HA'] = data['HA'].astype(int)
data['DIAB'] = data['DIAB'].astype(int)
data['TABAG'] = data['TABAG'].astype(int)
data['TABAG4'] = data['TABAG4'].astype(int)
```

#### Coeficiente de Determinação

```python
model = ols('IMC ~ SEXO', data = data).fit()
anova_result = sm.stats.anova_lm(model, typ=2)
print(anova_result)
```

```python
ss_total = anova_result['sum_sq'].sum()
ss_total
```

```python
ss_modelo = anova_result['sum_sq'][1]
ss_modelo
```

```python
r_squared = 1-(ss_modelo/ss_total)
r_squared
```
